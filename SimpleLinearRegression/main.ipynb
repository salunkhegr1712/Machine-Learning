# Simple Linear Regression model

## importing libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## importing the dataset

```python
data=pd.read_csv("Salary_Data.csv")

# print(data)

# now creating matrix of feature and dependent variable vector
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values

# we have to predict salary on basis of year of experience
```

## Splitting data in test and training set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1)
```

```python
print(len(xtrain),len(ytrain))
```

```python
print(len(xtest),len(ytest))
```

## Train the simple regression model on top of training dataset

```python
from sklearn.linear_model import LinearRegression

# now create a object for the linear LinearRegression
simpleRegressionModel=LinearRegression()

# and now using fit function to train the model on top of the training datset 
# this xtrain and y train order matter
simpleRegressionModel.fit(xtrain,ytrain)
```

### evaluting ans see predicted values from model

```python
# inorder to see what does model predict with xtest for test set we use predict function and
# it will cross check results with actual results that is ytest 

simpleRegressionModel.predict(xtest)
```

```python
print(ytest)
```

## Visulizing training dataset

```python
plt.scatter(xtrain,ytrain,color="red")

# now inorder to print regression line 
plt.plot(xtrain,simpleRegressionModel.predict(xtrain),color="black")
plt.title("salary vs experience")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```

# visulization for test dataset

```python
plt.scatter(xtest,ytest,color="red")

# now inorder to print regression line 
plt.plot(xtrain,simpleRegressionModel.predict(xtrain),color="black")
plt.title("salary vs experience")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```
